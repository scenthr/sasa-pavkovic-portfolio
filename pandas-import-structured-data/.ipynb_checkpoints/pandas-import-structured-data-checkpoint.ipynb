```python
import pandas as pd
```

```python
'''
With pandas most of the file types can be imported with few line of code. 
Here we show how to import a .csv, .json and texcel.
'''

births = pd.read_csv('../data/births.csv')
# Data taken from: https://www.kaggle.com/xvivancos/barcelona-data-sets#births.csv

iris = pd.read_json('../data/iris-json.json', orient='columns')
# Data taken from: https://www.kaggle.com/rtatman/iris-dataset-json-version

birthsxl =  pd.read_excel('../data/births.xlsx', sheet_name='births')
# births.csv from above converted into births.xlsx
```
