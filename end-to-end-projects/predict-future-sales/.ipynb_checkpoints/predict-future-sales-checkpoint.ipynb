### The question asked

We need to answer the question:
    - How many products will we sell in the next month, by product type and store? 
    
### The data provided

We have the daily sales of products per each store along side with other data that can be used to predict future sales. The data was provided by 1C company.

The 1С Company is an independent software developer, distributor and publisher company with its headquarters in Moscow, Russia. It develops, manufactures, licenses, supports, distributes and sells computer software, related services and video games. 1C is considered as a leading company in business software in Russia for its comprehensive business software suite 1C:Enterprise.

The data provided is separated in these data sets

    1. sales_train.csv - the training set. Daily historical data from January 2013 to October 2015.
    2. test.csv - the test set. You need to forecast the sales for these shops and products for November 2015.
    3. items.csv - supplemental information about the items/products.
    4. item_categories.csv - supplemental information about the items categories.
    5. shops.csv - supplemental information about the shops.
    
Data fields:

    1. ID - an Id that represents a (Shop, Item) tuple within the test set
    2. shop_id - unique identifier of a shop
    3. item_id - unique identifier of a product
    4. item_category_id - unique identifier of item category
    5. item_cnt_day - number of products sold. You are predicting a monthly amount of this measure
    6. item_price - current price of an item
    7. date - date in format dd/mm/yyyy
    8. date_block_num - a consecutive month number, used for convenience. January 2013 is 0, February 2013 is 1,..., October 2015 is 33
    9. item_name - name of item
    10. shop_name - name of shop
    11. item_category_name - name of item category
    
### Approach chosen

I decided to use the approach to transform the time series problem into a machine learning problem, hence then being able to apply some of the standard machine learning libraries for regression problems. The data includes the temporal component (*date* and *date_block_num* columns), so we will need to use the feature engineering in order to create features that capture that temporal component in form of lagged features. 

Not all of the attempts to understand the data and choose the best solution will be shown here.

### Data transformation

We inspect the test dataset as well and compare how many of the items and shops existed in the training data. For the items that are very new we will need to inpute the values. 

Then we will create one big matrix where one observation will hold infromation about sequence month (*date_block_num*), *shop_id* and *item_id* from both train and test sets. The aggregations will then be made and adde to the matrix.

Additional features will be added to the matrix in attempt to catch the temporal component in the data set. 

```python
# we transform the duplicate shop ids into unique ones

df = clear_shop_duplicates(df)
df_test = clear_shop_duplicates(df_test)
```

```python
def change_column_type(array, new_type):
    return array.astype(new_type)

def create_product_matrix(df,df_test):
    
    '''
    Given the train and test data produce a matrix with all the combinations
    of month, shop and item
    '''
    
    matrix = []
    cols = ['date_block_num','shop_id','item_id']
    for i in range(34):
        sales = df[df.date_block_num==i]
        matrix.append(np.array(list(product([i], sales.shop_id.unique(), sales.item_id.unique())), dtype='int16'))
    
    matrix = pd.DataFrame(np.vstack(matrix), columns=cols)
    matrix['date_block_num'] = change_column_type(matrix['date_block_num'],np.int8)
    matrix['shop_id'] = change_column_type(matrix['shop_id'],np.int8)
    matrix['item_id'] = change_column_type(matrix['item_id'],np.int16)
    matrix.sort_values(cols,inplace=True)

    # Group the number of items sold and add to matrix

    items = df.groupby(['date_block_num','shop_id','item_id']).item_cnt_day.sum()
    items = items.reset_index()
    items.columns = ['date_block_num','shop_id','item_id','item_cnt_month']
    
    cols = ['date_block_num','shop_id','item_id']
    matrix = pd.merge(matrix, items, on=cols, how='left')   
    matrix['item_cnt_month'] = matrix['item_cnt_month'].fillna(0)
    matrix['item_cnt_month'] = matrix['item_cnt_month'].clip(0,20)
    matrix['item_cnt_month'] = matrix['item_cnt_month'].astype(np.float16)
    
    ## adding the test information to matrix as 34. month - the one that we want to predict
    ## the data for
    
    df_test['date_block_num'] = 34
    df_test['date_block_num'] = df_test['date_block_num'].astype(np.int8)
    df_test['shop_id'] = df_test['shop_id'].astype(np.int8)
    df_test['item_id'] = df_test['item_id'].astype(np.int16)
    
    matrix = pd.concat([matrix, df_test], ignore_index=True, sort=False, keys=cols)
    matrix.fillna(0, inplace=True) 
    
    # Adding the item category id and item name to the matrix
    matrix = pd.merge(matrix, df_items, on=['item_id'], how='left')
    matrix['item_category_id'] = matrix['item_category_id'].astype(np.int8)
    matrix.drop(['item_name'], axis=1, inplace=True)

    matrix['month'] = matrix['date_block_num'] % 12

    return matrix
```

```python
# creating the needed matrix

matrix = create_product_matrix(df,df_test)
```

```python
def lag_feature(df, lags, col):
    
    '''
    Add lagged features for the item counts
    '''
    
    tmp = df[['date_block_num','shop_id','item_id',col]]
    for i in lags:
        shifted = tmp.copy()
        shifted.columns = ['date_block_num','shop_id','item_id', col+'_lag_'+str(i)]
        shifted['date_block_num'] += i
        df = pd.merge(df, shifted, on=['date_block_num','shop_id','item_id'], how='left')
    return df

# Added the lags for the feature we are trying to predict
matrix = lag_feature(matrix, [1,2,3,6,9,12], 'item_cnt_month')
```

```python
matrix.info()
```

```python
def add_group_mean_lag(df,groupby,agg_feature_name,lags,new_feature_name):
    
    '''
    Add the aggregate means in form of lags.
    
    Function also provides information about time needed to perform the
    operation.
    '''
       
    ts = time.time()
    
    loc = df.groupby(groupby).agg({agg_feature_name:['mean']})
    loc.columns = [new_feature_name]
    loc.reset_index(inplace=True)
    
    df = pd.merge(df,loc,on=groupby,how='left')
    df[new_feature_name] = df[new_feature_name].astype(np.float16)
    df = lag_feature(df, lags, new_feature_name)
    df.drop([new_feature_name], axis=1, inplace=True)    
    
    print('Elapsed: {} seconds'.format(time.time() - ts))
        
    return(df)
```

```python
# Lag for the last month mean average items sold
matrix = add_group_mean_lag(matrix,
                            ['date_block_num'],
                            'item_cnt_month',
                            [1],
                            'block_avg_item_cnt')
```

```python
# lags for month and item
matrix = add_group_mean_lag(matrix,
                            ['date_block_num', 'item_id']
                            ,'item_cnt_month',
                            [1,2,3,6,9,12],
                            'block_item_avg_item_cnt')
```

```python
# lags for month and shop
matrix = add_group_mean_lag(matrix
                            ,['date_block_num', 'shop_id']
                            ,'item_cnt_month'
                            ,[1,2,3,6,9,12]
                            ,'block_shop_avg_item_cnt')
```

```python
# lags for month and item category
matrix = add_group_mean_lag(matrix
                            ,['date_block_num', 'item_category_id']
                            ,'item_cnt_month'
                            ,[1,12]
                            ,'block_cat_avg_item_cnt')
```

```python
# lags for month and item category
matrix = add_group_mean_lag(matrix
                            ,['date_block_num', 'shop_id','item_category_id']
                            ,'item_cnt_month'
                            ,[1,11]
                            ,'block_shop_cat_avg_item_cnt')
```

```python
# We can add additional useful features

def add_agg_feature(df,groupby,aggby,aggfunc,new_feature_name):
    
    '''
    Add the aggregate columns based on the aggregate function provided
    '''
    
    loc = df[(df['date_block_num']!=34)].groupby(groupby).agg({aggby:[aggfunc]})
    loc.columns = [new_feature_name]
    loc.reset_index(inplace=True)
    
    df = pd.merge(df,loc,on=groupby,how='left')
    df[new_feature_name] = df[new_feature_name].astype(np.float16)
    df[new_feature_name].fillna(-1, inplace=True) 

    return df
```

```python
# Add period when the item was last sold (per shop)

matrix = add_agg_feature(matrix,['shop_id','item_id'],'date_block_num','max','shop_item_last_purchased')
```

```python
# Add period when the item was last sold

matrix = add_agg_feature(matrix,['item_id'],'date_block_num','max','item_last_purchased')
```

```python
# Add period when the item was first sold (per shop)

matrix = add_agg_feature(matrix,['shop_id','item_id'],'date_block_num','min','shop_item_first_purchased')
```

```python
# Add period when the item was first sold

matrix = add_agg_feature(matrix,['item_id'],'date_block_num','min','item_first_purchased')
```

```python
# finalize and store the matrix created

def finalize_and_store_matrix(df):
    
    '''
    Finalize the matrix created
    '''
    
    # we drop the first 12 months as we have used lags of max 12 months
    df = df[df.date_block_num > 11]
    
    # makes sense to fill the remaining values with 0 as there were no sales in the
    # previous lag for that combination of date_block_num, shop_id and item_id
    df.fillna(0,inplace=True)
    
    df.to_pickle('data/matrix_new.pkl')

    return df


# store the matrix to the disk for later use

matrix = finalize_and_store_matrix(matrix)
```

### Modelling 

We have a dataset prepared to start the modelling process. As we are trying to predict a cotinuous variable we can use regression to solve the problem

I will start with linear regression as a base model and then use some more advanced algorithms.

As we have a problem where temporal element is important we will split the data so that the last months are the ones in the test dataset and others in the training. 

```python
def createValidationSets(df, sets=3):
    '''
    By default creates 3 sets of train, validation and test data in time
    series manner (keeping the sequence of events in mind)
    '''
    
    cv_data = {} 
    
    for n in range(sets):
        
        period = df.date_block_num.max() - n
        period_m1 = df.date_block_num.max() - n - 1
        
        X_train = df[df.date_block_num < period_m1].drop(['item_cnt_month'], axis=1)
        Y_train = df[df.date_block_num < period_m1]['item_cnt_month']
        X_valid = df[df.date_block_num == period_m1].drop(['item_cnt_month'], axis=1)
        Y_valid = df[df.date_block_num == period_m1]['item_cnt_month']
        X_test = df[df.date_block_num == period].drop(['item_cnt_month'], axis=1)
        Y_test = df[df.date_block_num == period]['item_cnt_month']
                
        cv_data[n] = list([X_train, Y_train, X_valid, Y_valid, X_test])

    return cv_data
```

```python
cv_data_sets = createValidationSets(df)
```

```python
# Extreme Gradient Boosting fitting

# Overview of fold sizes

for n in range(len(cv_data_sets)):
    print(len(cv_data_sets[n][0]),len(cv_data_sets[n][1]),len(cv_data_sets[n][2]),len(cv_data_sets[n][3]))
```

```python
def runCVFoldsOnXGB(data_folds):
    
    '''
    Takes the time series folds and applies the XGBoost and stores the results
    in order to check for consistency
    '''

    model = XGBRegressor(
        max_depth=8,
        n_estimators=1000,
        min_child_weight=300, 
        colsample_bytree=0.8, 
        subsample=0.8, 
        eta=0.3,    
        seed=42
        )    
    
    for n in range(len(data_folds)):
        
        # this part is commented in order to speed up the process, hence only one fold is taken into account
        if n == 0:
        
            model.fit(
                data_folds[n][0], 
                data_folds[n][1], 
                eval_metric="rmse", 
                eval_set=[(data_folds[n][0], data_folds[n][1]), (data_folds[n][2], data_folds[n][3])], 
                verbose=True, 
                early_stopping_rounds = 3)

            Y_pred = model.predict(data_folds[n][2]).clip(0, 20)

            print("Fold {} ==>".format(n))

            print("Mean squared error: %.2f"        
                  % mean_squared_error(data_folds[n][3], Y_pred))

            print("Root mean squared error: %.2f"
                  % mean_squared_error(data_folds[n][3], Y_pred)**(1/2))

            print("Fold {} <==".format(n))
```

```python
# NOTE: This function might take a long time to finish

runCVFoldsOnXGB(cv_data_sets)
```

### Choosing the model

As the Xtreme Gradient Boosting Regressor is the best model out of the 2 with the MSE of 0.89, we can use this model for the predicitons.

The predictions need to be made on the test set that is hidden (usual form for the Kaggle competition), hence we can not check the results on the test data set, but just evaluate the model on the validation dataset above.

#### Computational resources

These datasets are relatively big, so fitting the model becomes time consuming. Alternative approaches to local computer resources are needed like creating a temporary powerful instance in the cloud (e.g. AWS) to fit the model and try different solutions.

#### Parameter tuning

In this problem we have the data with the temporal component so the validation sets need to be created manually by moving back in time by one month. This limits the parameter tuning somewhat, so it needs to be done in a specifically designed way.

#### Looking at RMSE

Looking at RMSE of 0.95 on the validation set looks very good,but we need to remember that most of the *shop_id*, *item_id* combinations have 0 or 1 items sold. So knowing that most of the predictions between 0 and 1 could be wrong, while the effect should be smaller for the items that sell a lot.

## Conclusions about the model

As it is now. the model could be used for the predictions of sales of bestseller items, but not so much the items that sell once per month. We would also need to decide how we want to aggregate the results if a prediction for store-item is 0.174456 units. 

Maybe a better approach would be to take a look at predicting the sales of total items from the stores only (or store-item category combination) as then the agggreate values should help make the regression problem easier.
